```python

import numpy as np
import re
import nltk
import pandas as pd
import nltk as nlp
nltk.download('stopwords')
import pickle
from nltk.corpus import stopwords
stopWords = set(stopwords.words('turkish'))
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python

# Metin temizleme fonksiyonu
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Noktalama işaretlerini kaldır
    text = text.lower()  # Küçük harfe çevir
    text = ' '.join(word for word in text.split() if word not in stop_words)  # Stopwords kaldır
    return text

# Eğitim ve test veri setlerini lokalden yükle
train_data = pd.read_csv("train.csv", encoding="ISO-8859-9")  # Eğitim veri seti
test_data = pd.read_csv("test.csv", encoding="ISO-8859-9")  # Test veri seti

# Metinleri temizle
train_data['cleaned_comment'] = train_data['comment'].apply(clean_text)
test_data['cleaned_comment'] = test_data['comment'].apply(clean_text)

# Temizlenmiş eğitim verisinin ilk birkaç satırını görüntüle
print("\nTemizlenmiş Eğitim Verisi:")
print(train_data[['cleaned_comment', 'Label']].head())

# Eğitim ve test verilerini ayrıştır
X_train = train_data["cleaned_comment"]
y_train = train_data["Label"]
X_test = test_data["cleaned_comment"]
y_test = test_data["Label"]

print("\nVeri Seti Boyutları:")
print(f"x_train: {X_train.shape}")
print(f"x_test: {X_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, precision_score, f1_score, recall_score
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import seaborn as sns

# Logistic Regression Pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),  # TF-IDF vektörizer
    ('clf', LogisticRegression())  # Logistic Regression modeli
])

# Modeli eğit
pipeline.fit(X_train, y_train)

print("\nModel Başarıyla Eğitildi!")

# 1. Çapraz Doğrulama Skorları (Cross Validation)
cv_scores = cross_val_score(pipeline, X_train, y_train, cv=10)
print("\nCV Ortalama Skoru: %.2f" % cv_scores.mean())

# 2. Test Verisi Üzerinde Tahmin
y_pred = pipeline.predict(X_test)

# 3. Sınıflandırma Raporu
cr = classification_report(y_test, y_pred)
print("\nSınıflandırma Raporu:\n", cr)

# 4. Eğitim ve Test Doğruluğu
train_accuracy = pipeline.score(X_train, y_train)
test_accuracy = pipeline.score(X_test, y_test)
print(f"\nEğitim Doğruluğu: {train_accuracy:.3f}")
print(f"Test Doğruluğu: {test_accuracy:.3f}")

# 5. Metriği Hesapla
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print(f"\nPrecision Skoru: {precision:.3f}")
print(f"Recall Skoru: {recall:.3f}")
print(f"F1 Skoru: {f1:.3f}")

```

```python
# Karışıklık Matrisi Çizimi
def plot_confusion_matrix(y_true, y_preds):
    conf_mat = confusion_matrix(y_true, y_preds)
    plt.figure(figsize=(6, 6))
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel("Tahmin Edilen")
    plt.ylabel("Gerçek")
    plt.title("Karışıklık Matrisi")
    plt.show()

# Çizimi çağır
plot_confusion_matrix(y_test, y_pred)
```

```python
# Kullanıcıdan yorum al ve tahminde bulun
def predict_single_comment(pipeline):
    """
    Kullanıcıdan alınan tek bir yorumu tahmin eder ve doğruluk oranlarını gösterir.
    """
    # Kullanıcıdan yorum al
    user_input = input("Lütfen bir yorum girin: ")
    
    # Yorumu temizle
    cleaned_comment = clean_text(user_input)
    
    # Tahmin ve olasılık hesaplama
    prediction = pipeline.predict([cleaned_comment])
    probabilities = pipeline.predict_proba([cleaned_comment])
    
    # Tahmin sonucunu yazdır
    if prediction[0] == 1:  # İyi (Positive)
        print(f"Yorum: {user_input} -> Olasılık: %{probabilities[0][1] * 100:.2f} Positive")
    else:  # Kötü (Negative)
        print(f"Yorum: {user_input} -> Olasılık: %{probabilities[0][0] * 100:.2f} Negative")

# Fonksiyonu çalıştır
predict_single_comment(pipeline)
```
